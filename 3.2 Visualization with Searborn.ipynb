```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("cars.csv")
df.head()
```

### Line plot

```python
x = np.arange(10)
y = np.sqrt(x)
plt.figure(figsize=(7,7))
# we use lineplot function to plot line graph with sea born library
sns.lineplot(x,y)
plt.show()
```

### Distplot aka Histogram

```python
plt.figure(figsize=(7,7))
sns.distplot(df["width"])
plt.show()
```

```python
width_mean = df["width"].mean()
width_median = df["width"].median()

plt.figure(figsize=(7,7))
sns.distplot(df["width"])

plt.axvline(width_mean, color="green")
plt.axvline(width_median, color="yellow")
plt.show()
```

###  Bar Plot / Count Plot

```python
#bar plot is called as count plot in seaborn
#count plot takes only one variable either x or y
plt.figure(figsize=(7,7))
sns.countplot(data=df, x="make")
plt.show()
```

```python
plt.figure(figsize=(7,7))
sns.countplot(data=df, x="make")
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(7,7))
sns.countplot(data=df, y="make")
plt.show()
```

```python
plt.figure(figsize=(7,7))
sns.countplot(data=df, x="body-style")
plt.show()
```

```python
#use hue when you wanna distribute or group your data further
plt.figure(figsize=(7,7))
sns.countplot(data=df, x="body-style",hue="fuel-type")
plt.show()
```

### HeatMap
Heat Maps are graphical representations of data that utilize color-coded systems. The primary purpose of Heat Maps is to better visualize the volume of locations/events within a dataset and assist in directing viewers towards areas on data visualizations that matter most.

```python
df.corr()
```

```python
#light color means positively coerelated and dark color means negatively corelated
cor = df.corr()
plt.figure(figsize=(7,7))
sns.heatmap(cor)
plt.show()
```

```python
#annot will display all the numbers values of each cell in the heat map
plt.figure(figsize=(9,7))
sns.heatmap(cor,annot=True) 
plt.show()
```

```python
#cmap is used to change the color of the heat map
sns.heatmap(cor,annot=True, cmap="viridis") 
plt.show()
```
